# Calibrate Synthetic Fields
Use this notebook to tweak gyre, wind, and Stokes parameters against target metrics before exporting configs.


```python
import numpy as np
import matplotlib.pyplot as plt
from driftcast.fields.gyres import GyreFieldConfig, gyre_velocity_field
cfg = GyreFieldConfig()
lon = np.linspace(-90, -20, 120)
lat = np.linspace(10, 60, 100)
lon_grid, lat_grid = np.meshgrid(lon, lat)
u, v = gyre_velocity_field(lon_grid, lat_grid)
speed = np.hypot(u, v)
plt.figure(figsize=(8,5))
plt.contourf(lon, lat, speed, levels=20)
plt.title("Synthetic Gyre Speed (m/s)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="m/s")
plt.show()
```
